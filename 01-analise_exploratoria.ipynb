Evolução de pedidos ao longo do tempo

```python
df['Data_Pedido'] = pd.to_datetime(df['Data_Pedido'])
pedidos_mes = df.groupby(df['Data_Pedido'].dt.to_period('M')).size()

plt.figure(figsize=(10,4))
pedidos_mes.plot(kind='line', marker='o')
plt.title('Número de pedidos por mês')
plt.xlabel('Mês')
plt.ylabel('Quantidade de pedidos')
plt.show()
```

Participação de cada tipo de fertilizante nas vendas

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Tipo_Fertilizante', order=df['Tipo_Fertilizante'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Distribuição de Tipos de Fertilizantes')
plt.show()
```

Receita por Região

```python
df['Receita'] = df['Quantidade_Toneladas'] * df['Preco_Ton_USD']

receita_regiao = df.groupby('Localizacao_Cliente')['Receita'].sum().sort_values()

plt.figure(figsize=(8,5))
receita_regiao.plot(kind='barh', color='skyblue')
plt.title('Receita Total por Região')
plt.xlabel('Receita (USD)')
plt.ylabel('Região')
plt.show()
```

Comparação de preço e custo de produção

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Custo_Producao_Ton_USD', y='Preco_Ton_USD', hue='Tipo_Fertilizante')
plt.title('Preço vs Custo de Produção por Tipo de Fertilizante')
plt.show()
```

Tempo de entrega vs status

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Status_Entrega', y='Tempo_Entrega_Dias')
plt.title('Tempo de Entrega por Status')
plt.show()
```

Correlação entre variáveis numéricas

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='Blues')
plt.title('Matriz de Correlação')
plt.show()

```
